```python
# Your parking gargage class should have the following methods:
# - takeTicket
    # - This should decrease the amount of tickets available by 1
    # - This should decrease the amount of parkingSpaces available by 1
# - payForParking
    # - Display an input that waits for an amount from the user and store it in a variable
    # - If the payment variable is not empty then (meaning the ticket has been paid) -> display a message to the user that their ticket has been paid and they have 15mins to leave
    # - This should update the "currentTicket" dictionary key "paid" to True
# -leaveGarage
    # - If the ticket has been paid, display a message of "Thank You, have a nice day"
    # - If the ticket has not been paid, display an input prompt for payment
    # - Once paid, display message "Thank you, have a nice day!"
    # - Update parkingSpaces list to increase by 1 (meaning add to the parkingSpaces list)
    # - Update tickets list to increase by 1 (meaning add to the tickets list)

# You will need a few attributes as well:
# - tickets -> list
# - parkingSpaces -> list
# - currentTicket -> dictionary

class parking_garage():
    def __init__(self):
        self.tickets_available = [1,2,3,4,5]
        self.parking_space_available = [1,2,3,4,5]
        self.current_ticket = {}
        self.ticket= 0
        self.space = 0

    def take_ticket(self):
        if self.parking_space_available:
            self.ticket = self.tickets_available.pop()
            self.space = self.parking_space_available.pop()
            self.current_ticket[self.ticket] = False
            print(f'ticket number = {self.ticket}')
        else:
            print("WE'RE FULL FIND ANOTHER LOT")
   

    def pay_for_parking(self):
        self.ticket = int(input('what is your ticket number?'))
        if self.ticket not in self.current_ticket.keys():
            print('Error. Ticket does not yet exist')
        elif self.current_ticket[self.ticket]:
            print(f'Ticket number {self.ticket} has been paid please exit.')
        elif self.ticket in self.current_ticket.keys():
            payment = input("The price of your ticket is: $10")
            if payment.isnumeric():
                if int(payment) == 10:
                    print('thank you for your payment')
                    print('you have 15 mins to leave the lot')
                    self.current_ticket[self.ticket] = True
                else:
                    print('that is the wrong amount. do better...')

            elif not payment.isnumeric():
                print(' stop playing and put a number')

    def leaving_garage(self):
        ticket_paid = int(input('what is your ticket number?'))

        if ticket_paid not in self.current_ticket:
            print(' this ticket is not in use')
        elif self.current_ticket[ticket_paid]:
            print(" Thank you for your business. Please come again.")
            self.tickets_available.append(self.ticket)
            self.parking_space_available.append(self.space)
            del self.current_ticket[self.ticket]
        elif self.current_ticket[ticket_paid] == False:
            print(' ticket has not been paid please return to kiosk.')
            
        

class Main():
    def __init__(self):
        self.garage = parking_garage()
    def instructions(self):
        print(f"""
        Welcome to Area 51 Garage!
        We currently have {len(self.garage.parking_space_available)} spaces available.
        Please choose an option:
        [1] Grab a ticket
        [2] Pay Ticket
        [3] Leave
        """)
    def run(self):
        self.instructions()
        while True:
            customer = input('Please choose an option: Grab a ticket/Pay Ticket/Leave')
            if customer == '1':
                self.garage.take_ticket()
            elif customer == '2':
                self.garage.pay_for_parking()
            elif customer == '3':
                self.garage.leaving_garage()
            # break


garage_final = Main()
garage_final.run()
```
